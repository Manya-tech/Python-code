```python
def insert(queue,data,front,rear,n):
    if front==0 and rear==n-1 or front==rear+1:
        print('queue is full')
    else:
        if front==-1:
            front=0
            rear=0
        elif rear==n-1 and front!=0:
            rear=0                          #circular queue
        else:
            rear+=1
        queue[rear]=data
    return front,rear

def delete(queue,front,rear,n):
    if front==-1:
        print('queue is empty')
    else:
        queue[front]=None
        if front==n-1 and rear>=0:
            front=0                            #circular queue
        elif front==rear:
            front=-1
            rear=-1
        else:
            front+=1
    return front,rear

def display(queue,front,rear,n):
    if front==-1:
        print('queue is empty')
    elif front<rear:
        for i in range (front,rear+1):
            print(queue[i], end='  ')
        print()
        print('Front=',front,'  Rear=',rear,'  Total accomodation=',n)
    elif front>rear:
        for i in range (front,n):
            print(queue[i],end='  ')
            
        for j in range (0,rear+1):
            print(queue[j], end='  ')
        print()
        print('Front=',front,'  Rear=',rear,'  Total accomodation=',n)

def main():
    q=[None,None,None,None,None,None,None,None,None]
    n=len(q)
    f=-1
    r=-1
    ch='y'

    print()
    while ch=='y':
        print('1. INSERT \
                2. DELETE \
                3. DISPLAY \
                4. QUIT')
        c=int(input('ENTER CHOICE:'))
        
        if c==1:
            da=int(input('ENTER DATA:'))
            f,r=insert(q,da,f,r,n)
            print('DONE \n')
            
        elif c==2:
            f,r=delete(q,f,r,n)
            print('DONE \n')
            
        elif c==3:
            dis=display(q,f,r,n)
            print()
        elif c==4:
            print('PROGRAM TERMINATED')
            break
        
        else:
            print('INVALID CHOICE')
        ch=input('GO ON?')
main()  
```
