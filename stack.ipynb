```python
def push(stk,data,maxstk):
    l=len(stk)
    if l==maxstk:
        print("OVERFLOW, STACK IS FULL")
    else:
        stk.append(data)
        print('DATA ADDED')

def pop(stk):
    if stk==[]:
        print("UNDERFLOW, STACK IS EMPTY")
    else:
        stk.pop()
        print("DATA DELETED")

def display(stk,maxstk):
    l=len(stk)
    print("MAXIMUM STACK=",maxstk)
    if l==0:
        print("STACK IS EMPTY")
    else: 
        print("STACK : ", end=' ')
        for i in range(l):
            print(stk[i], end=' ')
        print()
        print('STACK SUCCESSFULLY PRINTED')
s=[]
mxstk=5
ch='Y'
while ch=="Y":
    print()
    print('STACK OPERATIONS')
    print ("1. PUSH")
    print('2. POP')
    print("3. DISPLAY")
    print('4. QUIT')
    chh=int(input("ENTER YOUR CHOICE : "))
    if chh==1:
        d=int(input("ENTER THE ELEMENT TO PUSH :"))
        push(s,d,mxstk)
    elif chh==2:
        pop(s)
    elif chh==3:
        display(s,mxstk)
    elif chh==4:
        ch="N"
        print('PROGRAM TERMINATED')
    else:
        print('wrong choice')
```
